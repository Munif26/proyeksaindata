## **2. Kode untuk Load dan Inspeksi Data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (format UCR: kolom 0 = label, kolom 1-1500 = time series)
train_path = "SemgHandMovementCh2_TRAIN.txt"
test_path = "SemgHandMovementCh2_TEST.txt"

df_train = pd.read_csv(train_path, delim_whitespace=True, header=None)
df_test = pd.read_csv(test_path, delim_whitespace=True, header=None)

print("Bentuk Data Training (Baris, Kolom):", df_train.shape)  # (450, 1501)
print("Bentuk Data Testing (Baris, Kolom):", df_test.shape)    # (450, 1501)
print("\nContoh 5 baris pertama data training:")
display(df_train.head())

# Pisah label dan fitur
df_train[0] = df_train[0].astype(int)
df_test[0] = df_test[0].astype(int)

X_train = df_train.iloc[:, 1:]   # Kolom 1-1500: time series
y_train = df_train.iloc[:, 0]    # Kolom 0: label (1-6)
X_test = df_test.iloc[:, 1:]
y_test = df_test.iloc[:, 0]

print("Jumlah fitur (panjang time series):", X_train.shape[1])  # 1500
print("Unique labels train:", sorted(y_train.unique()))       # [1,2,3,4,5,6]
print("Unique labels test:", sorted(y_test.unique()))

# Cek missing values
print("Missing values TRAIN:", X_train.isnull().sum().sum())  # 0
print("Missing values TEST:", X_test.isnull().sum().sum())    # 0

# Distribusi kelas
print("\nDistribusi Kelas pada Training Set:")
print(pd.Series(y_train).value_counts().sort_index())
print("\nDistribusi Kelas pada Test Set:")
print(pd.Series(y_test).value_counts().sort_index())
```

## **3. Visualisasi**

```python
#Visualisasi Distribusi Label:
print ("Visualisasi Distribusi Label:")
plt.figure(figsize=(6,3))
sns.countplot(x=y_train)
plt.title("Distribusi Label (TRAIN)")
plt.xlabel("Label")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(6,3))
sns.countplot(x=y_test)
plt.title("Distribusi Label (TEST)")
plt.xlabel("Label")
plt.ylabel("Count")
plt.show()

print ("------------------------------------------------------------------------------------------------------------------------------------------------------------------------")




# Visualisasi Contoh Time Series:
unique_classes = np.unique(y_train)
plt.figure(figsize=(15, 10))
for i, c in enumerate(unique_classes):
    idx = np.where(y_train == c)[0][0]
    plt.subplot(2, 3, i + 1)
    plt.plot(X_train.iloc[idx])
    plt.title(f'Contoh Timeseries Kelas {c} ({["Cylindrical", "Hook", "Tip", "Palmar", "Spherical", "Lateral"][c-1]})')
    plt.xlabel('Timepoints')
    plt.ylabel('Nilai')
print ("Visualisasi Contoh Time Series:")
plt.tight_layout()
plt.show()
print ("------------------------------------------------------------------------------------------------------------------------------------------------------------------------")




# Visualisasi Rata-rata Time Series per Kelas:
print ("Visualisasi Rata-rata Time Series per Kelas:")
plt.figure(figsize=(12,5))
for label in unique_classes:
    mean_series = X_train[y_train == label].mean(axis=0)
    plt.plot(mean_series, label=f"Mean class {label}")
plt.legend()
plt.title("Rata-rata Time Series per Kelas (TRAIN)")
plt.xlabel("Time index")
plt.ylabel("Mean value")
plt.show()
```

# **C. DATA PREPROCESSING**

Tahap *data preprocessing* merupakan langkah penting dalam proses analisis data karena bertujuan untuk menyiapkan data agar dapat diproses secara optimal oleh algoritma machine learning. Pada dataset **SemgHandMovementCh2**, preprocessing difokuskan pada penyesuaian format data dan label agar sesuai dengan kebutuhan model klasifikasi *time series*, tanpa mengubah karakteristik utama sinyal sEMG.

Langkah pertama yang dilakukan pada tahap ini adalah **encoding label kelas**. Label asli pada dataset berada dalam rentang nilai **1 hingga 6**, yang masing-masing merepresentasikan enam jenis gerakan tangan. Namun, sebagian besar algoritma machine learning dan deep learning membutuhkan label dalam bentuk indeks numerik yang dimulai dari nol. Oleh karena itu, digunakan teknik **Label Encoding** untuk mengonversi label dari rentang 1–6 menjadi 0–5. Proses ini memastikan bahwa setiap kelas direpresentasikan secara konsisten dan dapat diproses oleh model tanpa menimbulkan kesalahan interpretasi.

Selanjutnya, data fitur yang semula berada dalam struktur *DataFrame* dikonversi ke dalam bentuk **array NumPy**. Konversi ini dilakukan karena sebagian besar proses komputasi numerik dan pemodelan machine learning di Python lebih efisien dan kompatibel dengan format NumPy. Dengan mengubah data ke bentuk array, proses pelatihan dan evaluasi model dapat dilakukan dengan lebih cepat dan stabil.

Pada tahap preprocessing ini, tidak dilakukan proses normalisasi tambahan karena dataset **SemgHandMovementCh2** telah disediakan dalam kondisi yang relatif terstandarisasi. Hal ini dapat diamati dari nilai rata-rata (*mean*) dan standar deviasi (*standard deviation*) sinyal yang berada pada rentang yang wajar. Informasi tersebut ditampilkan sebagai bagian dari pengecekan awal untuk memastikan bahwa data tidak memiliki skala yang ekstrem yang dapat memengaruhi kinerja model.

Sebagai tahap akhir, dilakukan pengecekan bentuk data (*shape*) untuk memastikan bahwa struktur data sudah sesuai dengan kebutuhan pemodelan. Bentuk data yang ditampilkan menunjukkan jumlah sampel dan panjang deret waktu, sehingga dapat dipastikan bahwa setiap sampel direpresentasikan sebagai *time series* dengan panjang yang konsisten. Berdasarkan hasil eksplorasi dan analisis distribusi kelas pada dataset **SemgHandMovementCh2**, dapat disimpulkan bahwa dataset ini memiliki distribusi kelas yang relatif **seimbang**. Teknik *Synthetic Minority Over-sampling Technique* (SMOTE) umumnya digunakan untuk mengatasi permasalahan ketidakseimbangan data dengan cara menambah sampel sintetis pada kelas minoritas. Namun, pada dataset ini, penerapan SMOTE tidak diperlukan karena tidak ada kelas yang secara signifikan kekurangan jumlah data. Selain itu, dataset **SemgHandMovementCh2** merupakan dataset *time series* biomedis, di mana struktur temporal dan kontinuitas sinyal memiliki peran penting. Penerapan SMOTE pada data *time series* dapat berisiko merusak pola temporal asli karena teknik ini pada dasarnya bekerja pada ruang fitur statis, bukan pada deret waktu. Oleh karena itu, mempertahankan data asli tanpa proses oversampling menjadi pilihan yang lebih tepat untuk menjaga keaslian pola sinyal.

Dengan mempertimbangkan keseimbangan distribusi kelas serta karakteristik data *time series* yang digunakan, keputusan untuk tidak menerapkan SMOTE dinilai tepat dan selaras dengan prinsip analisis data yang baik. Pendekatan ini membantu memastikan bahwa model belajar dari data asli yang representatif dan mampu melakukan generalisasi dengan lebih baik pada data uji.

## **1. Model 1-Nearest Neighbor (1-NN) dengan Jarak Euclidean**

Pada tahap pemodelan, algoritma K-Nearest Neighbor (KNN) digunakan sebagai model baseline untuk melakukan klasifikasi gerakan tangan berdasarkan sinyal sEMG. Konfigurasi yang dipilih adalah 1-Nearest Neighbor (1-NN) dengan metrik jarak Euclidean, di mana setiap data uji akan diklasifikasikan berdasarkan satu tetangga terdekatnya pada data latih. Pemilihan nilai k = 1 bertujuan untuk menangkap kemiripan paling langsung antar sinyal tanpa proses perataan dari banyak tetangga, sehingga model dapat merefleksikan sensitivitas tinggi terhadap pola sinyal individual.

Secara teknis, model dilatih menggunakan data fitur time series hasil preprocessing (X_train_np) dan label yang telah diencoding (y_train_enc). Setiap sampel direpresentasikan sebagai vektor berdimensi 1500 titik waktu, dan jarak Euclidean digunakan untuk menghitung kemiripan antar sinyal. Pendekatan ini mengasumsikan bahwa perbedaan nilai absolut pada setiap titik waktu memiliki kontribusi yang sama terhadap penentuan kelas, tanpa mempertimbangkan pergeseran temporal atau deformasi waktu.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 1-NN Euclidean
model_euclid = KNeighborsClassifier(n_neighbors=1, metric='euclidean')
model_euclid.fit(X_train_np, y_train_enc)
pred_euclid = model_euclid.predict(X_test_np)

print("Accuracy (1-NN Euclidean):", accuracy_score(y_test_enc, pred_euclid))
print("\nClassification Report:\n", classification_report(y_test_enc, pred_euclid))

# Confusion Matrix
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test_enc, pred_euclid), annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - 1-NN Euclidean")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

**1.1 Hasil Evaluasi Model 1-NN Euclidean**

- Berdasarkan hasil pengujian pada data uji, model 1-NN Euclidean menghasilkan akurasi sebesar 59,78%. Nilai ini menunjukkan bahwa model mampu mengklasifikasikan sekitar enam dari sepuluh sampel gerakan tangan dengan benar. Sebagai model baseline, performa ini memberikan gambaran awal mengenai tingkat kesulitan masalah klasifikasi gerakan tangan menggunakan sinyal sEMG mentah berbasis jarak statis.

- Laporan klasifikasi (classification report) menunjukkan variasi performa antar kelas. Beberapa kelas seperti kelas 4 (Spherical) dan kelas 5 (Lateral) memiliki nilai precision dan F1-score yang relatif tinggi, yang mengindikasikan bahwa pola sinyal pada kelas-kelas tersebut cukup konsisten dan mudah dikenali menggunakan jarak Euclidean. Sebaliknya, kelas seperti kelas 3 (Palmar) menunjukkan performa yang lebih rendah, baik dari sisi precision maupun recall, yang mengindikasikan adanya kemiripan pola sinyal dengan kelas lain atau variasi intra-kelas yang cukup tinggi.

**1.2 Analisis Confusion Matrix**

- Visualisasi confusion matrix memberikan wawasan lebih mendalam terkait kesalahan prediksi yang dilakukan model. Terlihat bahwa kesalahan klasifikasi sering terjadi antar kelas yang memiliki karakteristik sinyal sEMG yang mirip secara amplitudo dan bentuk gelombang. Hal ini mengindikasikan bahwa metrik Euclidean kurang mampu menangkap perbedaan pola temporal yang bersifat non-linear atau pergeseran fase waktu antar sinyal.

## **2. Model 1-Nearest Neighbor (1-NN) dengan Dynamic Time Warping (DTW)**

Sebagai pengembangan dari model baseline berbasis jarak Euclidean, penelitian ini mengimplementasikan 1-Nearest Neighbor (1-NN) dengan metrik Dynamic Time Warping (DTW). DTW merupakan metode pengukuran jarak yang dirancang khusus untuk data time series, karena mampu menyelaraskan dua sinyal dengan memperbolehkan peregangan dan pergeseran waktu. Pendekatan ini sangat relevan untuk sinyal sEMG, di mana pola gerakan otot dapat terjadi dengan kecepatan yang berbeda meskipun merepresentasikan gerakan yang sama.

Model ini diimplementasikan menggunakan pustaka tslearn, yang mendukung klasifikasi time series secara langsung tanpa harus mereduksi data menjadi fitur statis. Dengan menggunakan k = 1, setiap sampel data uji diklasifikasikan berdasarkan satu sampel data latih yang memiliki jarak DTW paling kecil, sehingga kemiripan pola temporal menjadi faktor utama dalam penentuan kelas.

```python
!pip install tslearn
```

```python
from tslearn.neighbors import KNeighborsTimeSeriesClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# ===============================
# 1. RESHAPE + DOWNSAMPLING
# ===============================
# dari 1500 → 300 timestep
X_train_ts = X_train_np.reshape((450, 1500, 1))[:, ::5, :]
X_test_ts  = X_test_np.reshape((450, 1500, 1))[:, ::5, :]

# ===============================
# 2. SUBSET DATA (AGAR CEPAT)
# ===============================
X_train_ts_small = X_train_ts[:100]
y_train_small = y_train_enc[:100]

X_test_ts_small = X_test_ts[:50]
y_test_small = y_test_enc[:50]

# ===============================
# 3. 1-NN DTW (OPTIMIZED)
# ===============================
model_dtw = KNeighborsTimeSeriesClassifier(
    n_neighbors=1,
    metric="dtw",
    metric_params={"sakoe_chiba_radius": 10}
)

model_dtw.fit(X_train_ts_small, y_train_small)
pred_dtw = model_dtw.predict(X_test_ts_small)

# ===============================
# 4. EVALUASI
# ===============================
print("Accuracy (1-NN DTW):", accuracy_score(y_test_small, pred_dtw))
print("\nClassification Report:\n", classification_report(y_test_small, pred_dtw))

plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test_small, pred_dtw),
            annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - 1-NN DTW")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

**2.1 Optimasi Data: Reshape, Downsampling, dan Subset**

- Karena kompleksitas komputasi DTW yang tinggi, terutama untuk sinyal panjang (1500 timestep), dilakukan downsampling dengan mengambil setiap titik ke-5 dari sinyal, sehingga panjang sinyal berkurang dari 1500 menjadi 300 timestep. Langkah ini bertujuan untuk menekan waktu komputasi tanpa menghilangkan pola utama sinyal secara signifikan.

- Selain itu, untuk keperluan eksperimen dan keterbatasan sumber daya komputasi, hanya digunakan subset data yang terdiri dari 100 sampel data latih dan 50 sampel data uji. Pendekatan ini bersifat eksploratif dan bertujuan untuk mengevaluasi potensi DTW dalam mengenali pola temporal sinyal sEMG, bukan untuk menghasilkan performa akhir yang sepenuhnya representatif terhadap keseluruhan dataset.

- Untuk meningkatkan efisiensi lebih lanjut, DTW dikonfigurasi dengan Sakoe-Chiba band (sakoe_chiba_radius = 10), yang membatasi ruang pencarian penyelarasan waktu sehingga mengurangi kompleksitas komputasi sekaligus membantu menghindari pencocokan yang tidak realistis.

**2.2 Hasil Evaluasi Model 1-NN DTW**

- Berdasarkan hasil pengujian pada subset data uji, model 1-NN DTW menghasilkan akurasi sebesar 90%, yang secara numerik jauh lebih tinggi dibandingkan model-model sebelumnya. Hal ini menunjukkan bahwa DTW sangat efektif dalam mengukur kemiripan sinyal sEMG berbasis pola temporal, terutama ketika pergeseran waktu antar sinyal diperbolehkan.

- Namun, hasil classification report menunjukkan adanya ketidakseimbangan distribusi kelas pada subset data uji, di mana hanya satu kelas yang muncul dalam data uji tersebut. **2.3 nalisis Confusion Matrix dan Validitas Hasil**

- Confusion matrix menunjukkan bahwa mayoritas prediksi jatuh pada satu kelas dominan, yang menjelaskan mengapa nilai akurasi tampak sangat tinggi. ## **3. (Model Random Forest:)**

Pemilihan dan Konfigurasi Model Random Forest.

Selain model berbasis jarak, penelitian ini juga menguji Random Forest Classifier sebagai pendekatan ensemble learning untuk klasifikasi gerakan tangan berbasis sinyal sEMG. Model ini diharapkan mampu menangkap hubungan non-linear pada data sinyal sEMG yang kompleks.

## **4. (Model Advanced: CNN untuk Time Series:**)

Sebagai pendekatan lanjutan (advanced model), penelitian ini mengimplementasikan Convolutional Neural Network (CNN) yang dirancang khusus untuk menangani data time series. CNN dipilih karena kemampuannya dalam mengekstraksi fitur lokal secara otomatis dari sinyal berurutan, seperti pola amplitudo, frekuensi lokal, dan transisi sinyal, tanpa memerlukan proses ekstraksi fitur manual. Pada konteks sinyal sEMG, CNN diharapkan mampu menangkap karakteristik kontraksi otot yang berbeda untuk setiap jenis gerakan tangan.

Sebelum pelatihan model, data time series hasil preprocessing di-reshape ke dalam format tiga dimensi (jumlah_sampel, panjang_sinyal, 1), di mana dimensi terakhir merepresentasikan satu kanal sinyal. Proses ini diperlukan agar data sesuai dengan input layer pada arsitektur Conv1D. Dengan panjang sinyal sebesar 1500 titik, setiap sampel menyediakan informasi temporal yang cukup kaya untuk dipelajari oleh model.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout

# Reshape untuk CNN (n_samples, n_timestamps, 1)
X_train_cnn = X_train_np.reshape((450, 1500, 1))
X_test_cnn = X_test_np.reshape((450, 1500, 1))

# Model CNN
model_cnn = Sequential([
    Conv1D(64, kernel_size=3, activation='relu', input_shape=(1500, 1)),
    MaxPooling1D(pool_size=2),
    Conv1D(128, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(6, activation='softmax')  # 6 kelas
])

model_cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model_cnn.fit(X_train_cnn, y_train_enc, epochs=10, validation_split=0.2, verbose=1)

# Prediksi
pred_cnn = np.argmax(model_cnn.predict(X_test_cnn), axis=1)
print("Accuracy (CNN):", accuracy_score(y_test_enc, pred_cnn))
print("\nClassification Report:\n", classification_report(y_test_enc, pred_cnn))

# Confusion Matrix
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test_enc, pred_cnn), annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - CNN")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

**4.1 Arsitektur dan Konfigurasi Model CNN**

- Arsitektur CNN yang digunakan terdiri dari dua lapisan Conv1D dengan jumlah filter bertambah (64 dan 128) dan ukuran kernel 3, yang berfungsi untuk mengekstraksi fitur lokal dari sinyal sEMG. Setiap lapisan konvolusi diikuti oleh MaxPooling1D untuk mereduksi dimensi temporal sekaligus mempertahankan fitur dominan. Setelah proses ekstraksi fitur, lapisan Flatten digunakan untuk mengubah representasi fitur menjadi vektor satu dimensi yang kemudian diproses oleh lapisan Dense.

- Untuk meningkatkan generalisasi dan mengurangi risiko overfitting, lapisan Dropout dengan tingkat 50% diterapkan sebelum lapisan output. Lapisan output menggunakan fungsi aktivasi softmax dengan 6 neuron, sesuai dengan jumlah kelas gerakan tangan. Model dikompilasi menggunakan optimizer Adam dan fungsi loss sparse categorical crossentropy, yang sesuai untuk kasus klasifikasi multi-kelas dengan label numerik. Proses pelatihan dilakukan selama 10 epoch dengan skema validation split sebesar 20% dari data latih.

**4.2 Hasil Pelatihan dan Evaluasi Model CNN**

- Berdasarkan hasil pelatihan, model CNN menunjukkan peningkatan akurasi pada data latih secara bertahap hingga mencapai sekitar 56%, namun performa pada data validasi dan data uji menunjukkan nilai yang jauh lebih rendah. Pada tahap pengujian menggunakan data test, model CNN menghasilkan akurasi sebesar 36%, yang merupakan performa terendah dibandingkan model-model sebelumnya.

- Visualisasi confusion matrix memperlihatkan distribusi kesalahan klasifikasi yang cukup signifikan.
```python
%%writefile app.py
import streamlit as st
import numpy as np
import joblib

# Load model dan encoder
model = joblib.load('saved_models/uas/best_model.pkl')
le = joblib.load('saved_models/uas/label_encoder.pkl')

# Mapping label ke nama gerakan (sesuai dataset: 1=Cylindrical, 2=Hook, dll.)
gerakan_map = {
    1: "Cylindrical",
    2: "Hook",
    3: "Tip",
    4: "Palmar",
    5: "Spherical",
    6: "Lateral"
}

st.title("Klasifikasi Gerakan Tangan berdasarkan Sinyal sEMG")
st.write("Upload file sinyal time series (.txt, 1 baris dengan 1500 nilai) untuk prediksi gerakan tangan.")
st.write("Dataset: SemgHandMovementCh2 (6 kelas gerakan). Model: 1-NN Euclidean (Akurasi ~60%).")

# Input: Upload file
uploaded_file = st.file_uploader("Upload file .txt", type="txt")

if uploaded_file is not None:
    try:
        # Load data (asumsikan 1 sampel, 1500 titik, format space-separated seperti dataset)
        data = np.loadtxt(uploaded_file).reshape(1, 1500)  # Reshape ke (1, 1500)

        # Prediksi
        pred_encoded = model.predict(data)
        pred_label = le.inverse_transform(pred_encoded)[0]  # Decode ke label asli (1-6)
        gerakan = gerakan_map[pred_label]  # Map ke nama gerakan

        st.success(f"Prediksi Gerakan: **{gerakan}** (Label: {pred_label})")

        # Visualisasi sinyal
        st.subheader("Visualisasi Sinyal Input")
        st.line_chart(data.flatten())

    except Exception as e:
        st.error(f"Error: Pastikan file berisi 1500 nilai numerik (space-separated). Detail: {str(e)}")

st.write("---")
st.write("**Penjelasan:** Model menggunakan jarak Euclidean untuk klasifikasi time series. Demo UAS mengikuti CRISP-DM.")
st.write("**Link GitHub:** [Tambahkan link repo kamu di sini]")
```
